# Exercicio 1 - REGRESSOR
Utilize o conjunto de dados utilizado no Exercício da Aula1: dados.txt

- Teste com 3 diferentes arquiteturas de redes multi-camadas (MLPs)
- Compare o erro MSE obtido com os modelos treinados com o erro MSE obtido com a regressão multivariada.


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Carrega os dados a partir do arquivo
data = pd.read_csv('dados.txt')

# Separa as features (X) do target (y)
X = data.iloc[:, :-1]  # Todas as colunas, exceto a última
y = data.iloc[:,-1:]   # Última coluna

# Normalizar os atributos entre 0 e 1
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Exemplo de exibição dos dados lidos
print("Dados de entrada (features):")
print(X_normalized)
print("\nDados de saída (target):")
print(y.head(3))
```

```python
from sklearn.model_selection import train_test_split

# Divide os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Exemplo de exibição dos conjuntos de treino e teste
print("Conjunto de treino:")
print("Features (X_train):")
print(X_train)
print("Target (y_train):")
print(y_train)

print("\nConjunto de teste:")
print("Features (X_test):")
print(X_test)
print("Target (y_test):")
print(y_test)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

# Inicialização das arquiteturas de MLPs
mlp1 = MLPRegressor(hidden_layer_sizes=(10,), activation='identity', solver='lbfgs', max_iter=500, random_state=42)
mlp2 = MLPRegressor(hidden_layer_sizes=(5,), activation='relu', solver='lbfgs', max_iter=500, random_state=42)
mlp3 = MLPRegressor(hidden_layer_sizes=(1,), activation='logistic', solver='lbfgs', max_iter=500, random_state=42)

# Treinamento das MLPs
mlp1.fit(X_train, y_train)
mlp2.fit(X_train, y_train)
mlp3.fit(X_train, y_train)

# Previsões das MLPs nos dados de teste
mlp1_pred = mlp1.predict(X_test)
mlp2_pred = mlp2.predict(X_test)
mlp3_pred = mlp3.predict(X_test)

# Inicialização e treinamento da regressão multivariada
regression = LinearRegression()
regression.fit(X_train, y_train)

# Previsão da regressão multivariada nos dados de teste
regression_pred = regression.predict(X_test)

# Cálculo do erro MSE para cada modelo
mlp1_mse = mean_squared_error(y_test, mlp1_pred)
mlp2_mse = mean_squared_error(y_test, mlp2_pred)
mlp3_mse = mean_squared_error(y_test, mlp3_pred)
regression_mse = mean_squared_error(y_test, regression_pred)

# Exibição dos resultados
print("Erro MSE da MLP1:", mlp1_mse)
print("Erro MSE da MLP2:", mlp2_mse)
print("Erro MSE da MLP3:", mlp3_mse)
print("Erro MSE da regressão multivariada:", regression_mse)
```

Após analisar diversas combinações, deixamos as 3 acima como exemplos, pois alcançaram resultados próximos à regressão linear, sendo que a primeira teve um resultado muito semelhante, a segunda foi ligeiramente melhor e a terceira ficou um pouco acima.

# Exercicio 2 - CLASSIFICADOR

## Multilayer Perceptron para classificação do dataset Wine
Neste exercício utilizaremos a base de dados Wine, que consiste em 178 exemplos de dimensionalidade 13, separados em 3 classes. Os atributos dessa base de dados não estão normalizados, apresentando intervalos variados de valores.

- Carregue a base de dados wine e normalize entre 0 e 1 o valor dos seus atributos

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler

# Carregar a base de dados wine
wine = load_wine()

# Obter as features (X) e os targets (y)
X = wine.data
y = wine.target

# Normalizar os atributos entre 0 e 1
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Exemplo de exibição dos dados normalizados
print("Dados normalizados:")
print(X_normalized)
```

---
- Agora divida a base em conjunto de teste e treino. Utilize 20% da base para teste e 80% para treino.

```python
from sklearn.model_selection import train_test_split

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Exemplo de exibição dos conjuntos de treino e teste
print("Conjunto de treino:")
print("Features de treino (X_train):")
print(X_train)
print("Targets de treino (y_train):")
print(y_train)

print("\nConjunto de teste:")
print("Features de teste (X_test):")
print(X_test)
print("Targets de teste (y_test):")
print(y_test)
```

---
- Treine 3 classificadores MLP (com `solver='sgd'`), variando o número de neurônios na camada escondida. Reporte o score dos classificadores nos conjuntos de treino e teste.
  - Parâmetros para  alterar: hidden_layer_sizes, max_iter,  tol, learning_rate_init

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV


# Definir o espaço de busca dos parâmetros
param_grid = {
    'hidden_layer_sizes': [(10,), (50,), (100,)],
    'max_iter': [200, 500, 1000],
    'tol': [1e-3, 1e-4, 1e-5],
    'learning_rate_init': [0.001, 0.01, 0.1]
}

# Criar o objeto GridSearchCV
grid_search = GridSearchCV(
    estimator=MLPClassifier(hidden_layer_sizes=20, solver='sgd', alpha=0.001,learning_rate='adaptive', max_iter=1000),
    param_grid=param_grid,
    cv=5
)

# Realizar a busca pelos melhores hiperparâmetros
grid_search.fit(X_train, y_train)

# Obter os melhores hiperparâmetros encontrados
best_params = grid_search.best_params_

print("Melhores hiperparâmetros:")
for param, value in best_params.items():
    print(f"{param}: {value}")

# Avaliar o score do modelo com os melhores hiperparâmetros nos conjuntos de treino e teste
train_score = grid_search.score(X_train, y_train)
test_score = grid_search.score(X_test, y_test)

print("\nScore de treino com os melhores hiperparâmetros:", train_score)
print("Score de teste com os melhores hiperparâmetros:", test_score)

```

---
- Selecione uma amostra pequena de tamanho 10 do conjunto de testes. Selecione o classificador que apresentou maior acurácia no conjunto de testes e calcule, para cada elemento da amostra, a classe esperada, a classe obtida e a probabilidade estimada de cada classe

```python
# Melhor modelo com os hiperparâmetros selecionados
best_model = MLPClassifier(hidden_layer_sizes=(10,), learning_rate_init=0.01, max_iter=500, tol=0.0001)
best_model.fit(X_train, y_train)

# Selecionar uma amostra de tamanho 10 do conjunto de testes
sample_size = 10
X_sample = X_test[:sample_size]
y_sample = y_test[:sample_size]

# Calcular a classe esperada, a classe obtida e a probabilidade estimada de cada classe para cada elemento da amostra
expected_classes = y_sample
predicted_classes = best_model.predict(X_sample)
predicted_probabilities = best_model.predict_proba(X_sample)

# Exibir os resultados para cada elemento da amostra
for i in range(sample_size):
    print(f"Elemento {i+1}:")
    print("Classe esperada:", expected_classes[i])
    print("Classe obtida:", predicted_classes[i])
    print("Probabilidade estimada de cada classe:")
    for j, prob in enumerate(predicted_probabilities[i]):
        print(f"Classe {j}: {prob}")
    print("------------------------")
```

```python
# Pior modelo com os hiperparâmetros selecionados
worst_model = MLPClassifier()
worst_model.fit(X_train, y_train)


# Calcular a classe esperada, a classe obtida e a probabilidade estimada de cada classe para cada elemento da amostra
expected_classes = y_sample
predicted_classes = worst_model.predict(X_sample)
predicted_probabilities = worst_model.predict_proba(X_sample)

# Exibir os resultados para cada elemento da amostra
for i in range(sample_size):
    print(f"Elemento {i+1}:")
    print("Classe esperada:", expected_classes[i])
    print("Classe obtida:", predicted_classes[i])
    print("Probabilidade estimada de cada classe:")
    for j, prob in enumerate(predicted_probabilities[i]):
        print(f"Classe {j}: {prob}")
    print("------------------------")
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

# Realizar 10-fold cross-validation e calcular a acurácia média e o desvio padrão
scores = cross_val_score(best_model, X_normalized, y, cv=10)

# Exibir a acurácia média e o desvio padrão
mean_accuracy = scores.mean()
std_accuracy = scores.std()

print("Acurácia média:", mean_accuracy)
print("Desvio padrão:", std_accuracy)
```
